+ shallow - 얕은복사
+ deepcopy - 깊은복사

### 1. 그냥 할당

```python
lst = [1, 2, 3]
```

```python
a = lst
a[0] = 0
print(a, lst)
```

```python
a = lst.copy()
a[0] = 0
print(a, lst)
```

### 2. copy

```python
lst_lst = [[1, 2, 3], [4, 5, 6]]
```

```python
a = lst_lst.copy()
a[0][0] = 0
print(a, lst_lst)
```

### 3. deepcopy

```python
import copy
a = copy.deepcopy(lst_lst)
a[0][0] = 0
print(a, lst_lst)
```

### Pivot Table

[Pivot Table 다양하게 사용하기](https://www.analyticsvidhya.com/blog/2020/03/pivot-table-pandas-python/)

```python
# Type1별로 Legendary 의 비율을 보여주는 피벗 테이블
pd.pivot_table(pokemon, index="Type 1", values="Legendary").sort_values(by=["Legendary"], ascending=False)
```
