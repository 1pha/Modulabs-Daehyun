```python
import pandas as pd
```

```python
rating_cols = ['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_csv('../ml-1m/ratings.dat',
                      sep='::', header=None, names=rating_cols)
ratings
```

```python
ratings = ratings[rating_cols[:-1]]
ratings = ratings[ratings.rating >= 3]
```

```python
user_dict = {v: k for k, v in enumerate(ratings.user_id.unique())}
movie_dict = {v: k for k, v in enumerate(ratings.movie_id.unique())}

tmp_user_dict = ratings.user_id.map(user_dict.get).dropna()
ratings['user_id'] = tmp_user_dict

tmp_movie_dict = ratings.movie_id.map(movie_dict.get).dropna()
ratings['movie_id'] = tmp_movie_dict
```

```python
ratings.user_id.max()
```

```python
ratings.movie_id.max()
```

```python
from scipy.sparse import csr_matrix

csr_rating = csr_matrix((ratings.rating, (ratings.user_id, ratings.movie_id)),
                       shape=(ratings.user_id.max()+1, ratings.movie_id.max()+1))
```

```python
from implicit.als import AlternatingLeastSquares

als_model = AlternatingLeastSquares()
```

```python
csr_t = csr_rating.T
als_model.fit(csr_t)
```

```python
movie_dict_inverse = {v: k for k, v in movie_dict.items()}
```

```python
tmp_movie_dict_inverse = ratings.movie_id.map(movie_dict_inverse.get).dropna()
```

```python
ratings.movie_id = tmp_movie_dict_inverse
```

```python
def get_correct_title(idx):
    
    tag1 = movie_dict_inverse.get(idx)
    if tag1:
        return movies[movies.movie_id == tag1].title.values[0]
    
    else:
        return None
```

```python
ratings['title'] = ratings.movie_id.map(get_correct_title)
```
